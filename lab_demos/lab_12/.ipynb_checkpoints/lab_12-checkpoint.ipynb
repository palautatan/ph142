### Vocabulary
<font color="indigo">
<b>Causal effect v. association</b> An association does not imply causal. The regression model will tell you about association between two variables.

We call these variables <b>explanatory and response variables</b>, unfortunately. We have the <b>X</b>planatory and the response as Y. 

Once we have a linear model based on our data, we can calculated <b>predicted or fitted values</b>. We can subtract our predicted values from our observed values to calculate our <b>residuals</b>.
</font>

## Regression Test

<font color="indigo">
We are testing hypotheses about the <b>slope</b> and the <b>intercept</b>.
</font>

### Assumptions
<font color="indigo">
These assumptions are different than the ones in the book. <b>To test these assumptions, we look at graphs.</b>
<ul>
    <li>$x, y$ linear in scatterplot</li>
    <li>Residuals are normally distributed.</li>
    <li>Independent observations</li>
    <li>Standard deviation of the response variables are the same for all values of x</li>
    </ul>
</font>

### The test itself
<font color="indigo">
These are assumptions for $\beta_1$.    

$H_o: \, \beta_1 = 0$
<br>
$H_1: \, \beta_1 \neq 0$

```r
library(broom)
library(ggplot2)
library(dplyr)
library(tidyr)
```

```r
# GIVES ASSOCIATED "REGRESSION VALUES" WITH EACH COORDINATE
augment_Boston <- augment(lm_Boston)
```

```r
head(augment_Boston)
```

<font color="indigo">
Here's the fitted model plotted on top of the scatterplot.
</font>

```r
ggplot(augment_Boston, aes(nox, medv)) +
    geom_segment(aes(xend = nox, yend = .fitted), color="darkgrey") +
    geom_point(col="midnightblue") +
    ggtitle("Look at the residuals") +
    geom_smooth(method = "lm", se = F, col="midnightblue")
```

```r
ggplot(augment_Boston, aes(nox, medv)) +
    geom_point(col="darkgrey") +
    ggtitle("Look at the line") +
    ylab("Median House Price") +
    xlab("Air Pollution (Nitrogen Oxide)") +
    geom_smooth(method = "lm", se = F, col="midnightblue")
```

```r
ggplot(augment_Boston, aes(sample = .resid)) + 
  geom_qq() + 
  geom_qq_line(col="darkgrey")
```

```r
ggplot(augment_Boston, aes(y = .resid, x = .fitted)) + 
  geom_point(col="midnightblue") +
  geom_hline(aes(yintercept = 0)) +
  labs(y = "Residuals", x = "Fitted values", title = "(c) Fitted vs. residuals") 
```

```r
reshape <- augment_Boston %>% dplyr::select(.resid, medv) %>%
  gather(key = "type", value = "value", medv, .resid)

ggplot(reshape, aes(y = value)) +
  geom_boxplot(aes(fill = type)) +
  ggtitle("Look at the variation in residuals and in the response") +
  scale_fill_manual(values = c("darkgrey", "gold"))
```
